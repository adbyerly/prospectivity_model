# Part 1. import packages and processed data


```python
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt

data_path = "/Users/adbyerly/prospectivity_model/data/processed/"

grid = gpd.read_file(data_path + 'grid.geojson')
```

```python
print(grid.head())
```

# Part 2. spatial analysis

```python
# examine geochem distribution in mine vs. non-mine cells

mine_cells = grid[grid['gold'] == 1]              # all geochemical data in grid, but only for cells WITH mines
background_cells = grid[grid['gold'] == 0]        # all geochemical data in grid, but only for cells WITHOUT mines

print(mine_cells.head())
print(background_cells.head())
```

```python
# plot geochemical gold distributions for cells with mines and cells without

print(mine_cells['auppb_na'].describe())

plt.hist(mine_cells['auppb_na'], range = (0, 20), bins = 10)
plt.xlim(0, 20)
plt.show()

```

```python
print(background_cells['auppb_na'].describe())

plt.hist(background_cells['auppb_na'], range = (0, 20), bins = 'auto')
plt.xlim(0, 20)
plt.show()
```
